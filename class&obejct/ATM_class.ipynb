```python
class Atm:
    
    #constructor(special function)->superpower-> it gets excuted as soon as we make object
    def __init__(self):

        self.pin = ''
        self.balance = 0
        self.menu()

    def menu(self):
        user_input = input("""
        Hi how can i Help you ?
        1. Press 1 to create pin \n
        2. Press 2 to change pin \n
        3. Press 3 to check Balance \n
        4. Press 4 to withdraw \n
        5. Press Anything else to Exit\n""")

        if user_input == '1':
            #create
            self.create_pin()
        elif user_input =='2':
            #change
            self.change_pin()
        elif user_input =='3':
            self.check_balance()
            pass
        elif user_input=='4':
            self.withdraw_balance()
            pass
        else:
           exit()

    def create_pin(self):

        user_pin = input("enter your pin:")
        self.pin = user_pin

        user_balance = int(input("enter your Account Balance:"))
        self.balance = user_balance

        print("Pin created successfully")
        self.menu()

    def change_pin(self):
        old_pin = input("enter old pin:")

        if old_pin == self.pin:
            new_pin = input("enter new pin:")
            self.pin = new_pin
            print("pin updated successfully")
            self.menu()
        else:
            print("Wrong pin")
            self.menu()
    
    def check_balance(self):
        user_pin = input("enter pin:")

        if user_pin == self.pin:
            print("Your balance is:",self.balance)
        else:
            print("Wrong pin nikal")
        self.menu()
    
    def withdraw_balance(self):
        user_pin = input("enter pin:")

        if user_pin == self.pin:
            amount = int(input("enter amount to withdraw:"))
            if amount<= self.balance:
                self.balance = self.balance-amount
                print("Withdraw successfull \n remaning balance is:",self.balance)
            else:
                print("garib")
        else:
            print("wrong pin")
        self.menu()
        


```

```python
import random

class FlashCard:
    def __init__(self):
        self.fruits = {
            "Banana": "yellow",
            "Strawberries": "pink",
            "Watermelon": "green",
            "Apple": "red",
            "Grapes": "green"
        }

    def start_quiz(self):
        print("Welcome to fruit quiz")

        while True:
            fruit, color = random.choice(list(self.fruits.items()))

            print(f"What is the color of {fruit}")
            user_answer = input().lower()

            if user_answer == color:
                print("Correct answer")
            else:
                print("Wrong answer")

            choice = input("Enter 0, if you want to play again: ")
            if choice != "0":
                break
```

```python
fc = FlashCard()
fc.start_quiz()
```
